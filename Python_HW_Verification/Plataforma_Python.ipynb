```python
"""
  Formas utiles de trabajar:

    Crear una matriz 4*4 de ceros:
    a = np.zeros((4, 4), dtype=int)

    Insertar una columna de '1' en la posicion 0 de la matriz:
    a = np.insert(a, 0,np.ones((1,4), dtype=int), axis=1)

    Indexar un arreglo:
    a = A[0,:]  # Accedo a la primer fila de A
    a = A[:,0]  # Accedp a la primer col. de A

    Transponer un vector:
    a = a.T = a.transpose()

    Espejar una matriz (espeja por filas, con axis=1 espeja por columnas):
    A = np.flip()

"""

# Dimensiones de las matrices

N = 4   # Filas
M = 4   # Col.

# Matrices de entrada
A = np.ones((N,M), dtype=int)
B = np.ones((N,M), dtype=int)

# Matrices de salida
A1 = np.zeros((N, 2*M-1), dtype=int)
B1 = np.zeros((2*N-1, M), dtype=int)

print('Las matrices de entrada son:')
print('A:\n', A,'\nB:\n', B, '\n\n')

for i in range(N):
    a = A[i,:]                                  # Obtengo la primer fila
    a = np.insert(a, 0, np.zeros(M-1-i))        # Le inserto los ceros adelante   
    a = np.insert(a, 2*M-1-i, np.zeros(i))      # Le inserto los ceros atras 
    A1[i,:] = a                                 # Guardo en la matriz de salida

    b = B[:,i]                                  # Obtengo la primer col.
    b = b.T                                     # La transpongo para trabajarla como vector y le hago las mismas operaciones que a A
    b = np.insert(b, 0, np.zeros(N-1-i))          
    b = np.insert(b, 2*N-1-i, np.zeros(i))     
    b = b.T                                     # La vuelvo a transponer para guardarla en la matriz de salida
    B1[:,i] = b

print('Las matrices de salida son:')
print('A1:\n', A1, '\nB1:\n', B1)
print()

# Se rellenan las matrices dan vuelta para multiplicar mas facilmente los vectores

A1 = np.insert(A1, 0, np.zeros((N,M)), axis=1)
#print('La matriz A ampliada es:')
#print('A1:\n', A1)
#print()

B1 = np.insert(B1, 0, np.zeros((N,M)), axis=0)
#print('La matriz B ampliada es:')
#print('B1:\n', B1)
#print()

print("\nmatriz A1 dada vuelta en columnas:\n")
print(np.flip(A1, axis=1))
print()

#print("\nmatriz B1 transpuesta:\n")
#print(B1.T)
#print()

print("\nmatriz B1 transpuesta dada vuelta en columnas:\n")
print(np.flip(B1.T, axis=1))
print()

A1 = np.flip(A1, axis=1)
B1 = np.flip(B1.T, axis=1)


#sysmult.SystolicMult(A1,B1)

gensysmult.GeneralSystolicMult(A1,B1)
```

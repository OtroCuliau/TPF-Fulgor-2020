```python
import numpy as np
#import SystolicMult as sysmult
#import GeneralSystolicMult as gensysmult
import InputGenerator as inputgen
import CounterGenSysMult as countsysmult
```

```python
# Dimensiones de las matrices

N = 4   # Filas
M = 4   # Col.

# Matrices de entrada

A = np.ones((N,M), dtype=int)
B = np.ones((N,M), dtype=int)
C = 10*A
print(C)
D = C

premult_A = np.ones((N,2*M), dtype=int)
postmult_B = np.ones((2*N,M), dtype=int)

for i in range(N):
    print(i)
    print(A[i])
    A_aux = A[i,:]
    B_aux = B[i,:]
    A_aux = np.insert(A_aux,0,C[i])
    B_aux = np.insert(B_aux,0,D[i])
    
    premult_A[i,:] = A_aux
    postmult_B[:,i] = B_aux

print('premult_A',premult_A)
print('postmult_B: ',postmult_B)
```

```python
[A1,B1] = inputgen.InputGenerator(premult_A,postmult_B)

print('premult_A_stream:\n',A1)
print('postmult_B_stream:\n',B1)

countsysmult.CounterGenSysMult(A1,B1)
```
